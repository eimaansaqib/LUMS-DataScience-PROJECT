```python
all_dfs = [pc, ps3, ps4, ps5, xbox360, xboxone]
```

```python
pc = pc[pc["error"].isnull()]
ps3 = ps3[ps3["error"].isnull()]
ps4 = ps4[ps4["error"].isnull()]
ps5 = ps5[ps5["error"].isnull()]
xbox360 = xbox360[xbox360["error"].isnull()]
xboxone = xboxone[xboxone["error"].isnull()]
```

```python
def process(txt):
    return re.sub('[\[\]\']', '', txt)
```

```python
cols = ['genres','developers','also_on','also_on_url']
for col in cols:
    pc[col][pc[col].notnull()] = pc[col][pc[col].notnull()].map(process)
    ps3[col][ps3[col].notnull()] = ps3[col][ps3[col].notnull()].map(process)
    ps4[col][ps4[col].notnull()] = ps4[col][ps4[col].notnull()].map(process)
    ps5[col][ps5[col].notnull()] = ps5[col][ps5[col].notnull()].map(process)
    xbox360[col][xbox360[col].notnull()] = xbox360[col][xbox360[col].notnull()].map(process)
    xboxone[col][xboxone[col].notnull()] = xboxone[col][xboxone[col].notnull()].map(process)
```

```python
pc.rename(columns={"name":"title"}, inplace=True)
ps3.rename(columns={"name":"title"}, inplace=True)
ps4.rename(columns={"name":"title"}, inplace=True)
ps5.rename(columns={"name":"title"}, inplace=True)
xbox360.rename(columns={"name":"title"}, inplace=True)
xboxone.rename(columns={"name":"title"}, inplace=True)
```

```python
merge_cols = ['title','release_date']
```

```python
def join_str(x):
    x=x.reset_index(drop=True)
    if len(x)<2:
        return x[0]
    if type(x[0])==str and type(x[1])==str:
        if x[0]==x[1]:
            return x[0]
        return ','.join(x)
    elif type(x[0])==str:
        return x[0]
    elif type(x[1])==str:
        return x[1]
    else:
        return x[0]
```

```python
def merge_dfs(df1, df2,merge_col=merge_cols):
    df3 = df1.merge(df2,on=list(df1.columns), how='outer').groupby(merge_col, as_index=False).agg(
        {'publishers':lambda x:join_str(x),'developers':lambda x:join_str(x),'rating':lambda x:join_str(x),
         'metascritic_score':'max','users_score':'max','genres':'sum'})
    return df3
```

```python
primary = merge_dfs(merge_dfs(merge_dfs(merge_dfs(merge_dfs(pc, ps3),ps4),ps5),xbox360),xboxone)
```

```python
month_map={'Jan':'1','Feb':'2','Mar':'3','Apr':'4',"May":'5','Jun':'6',
           'Jul':'7','Aug':'8','Sep':'9','Oct':'10','Nov':'11','Dec':'12',
          'January':'1','February':'2','March':'3','April':'4',"May":'5','June':'6',
           'July':'7','August':'8','September':'9','October':'10','November':'11','December':'12'}
```

```python
def str_to_date(date_):
    stop = ['TBA', 'Canceled', 'Q1', 'Q2']
    for word in stop:
        if word in date_:
            return date_
    parts = date_.split()
    parts_notempty = []
    for part in parts:
        if len(part)>0:
            parts_notempty.append(part.split(',')[0])
    if len(parts_notempty)==1:
        return parts_notempty[0]
    if parts_notempty[0]=='1996':
        print(date_)
    parts_notempty[0] = month_map[parts_notempty[0]]
    for i in range(2):
        if len(parts_notempty[i])<2:
            parts_notempty[i] = '0' + parts_notempty[i]
    date_str = '/'.join(parts_notempty)
    try:
        new_date = datetime.strptime(date_str, '%m/%d/%Y').date()
        new_date = new_date.strftime('%Y/%m/%d')
    except:
        new_date = datetime.strptime(date_str, '%m/%Y').date()
        new_date = new_date.strftime('%Y/%m')
    return new_date
```

```python
primary['release_date']=primary['release_date'].apply(lambda x: str_to_date(x))
```

```python
indices = []
title_counts = primary['title'].value_counts()
for name in primary['title']:
    if title_counts[name]>1:
        names = primary[primary['title']==name]['release_date']
        index = list(names.index)
        maximum = np.argmax(names)
        index.pop(maximum)
        indices += index
indices = list(set(indices))
primary = primary.drop(indices, axis=0)
primary = primary.reset_index()
```
